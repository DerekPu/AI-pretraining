## Back to square one

[â¬†](#Sections)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)

for i in range(1, 4):
    plt.plot(x, i * x**2, label='Group %d' % i)

plt.legend(loc='best')
plt.show()
```

<br>
<br>

## Let's get fancy

[â¬†](#Sections)

```python
x = np.arange(10)

for i in range(1, 4):
    plt.plot(x, i * x**2, label='Group %d' % i)

plt.legend(loc='best', fancybox=True, shadow=True)
plt.show()
```

## Thinking outside the box

[â¬†](#Sections)

```python
fig = plt.figure()
ax = plt.subplot(111)

x = np.arange(10)

for i in range(1, 4):
    ax.plot(x, i * x**2, label='Group %d' % i)

ax.legend(loc='upper center', 
          bbox_to_anchor=(0.5,  # horizontal
                          1.15),# vertical 
          ncol=3, fancybox=True)
plt.show()
```

```python
fig = plt.figure()
ax = plt.subplot(111)

x = np.arange(10)

for i in range(1, 4):
    ax.plot(x, i * x**2, label='Group %d' % i)

ax.legend(loc='upper center', 
          bbox_to_anchor=(1.15, 1.02),
          ncol=1, fancybox=True)
plt.show()
```

## I love when things are transparent, free and clear

[â¬†](#Sections)

```python
x = np.arange(10)

for i in range(1, 4):
    plt.plot(x, i * x**2, label='Group %d' % i)

plt.legend(loc='upper right', framealpha=0.1)
plt.show()
```

## Markers -- All good things come in threes!

[â¬†](#Sections)

```python
from itertools import cycle
x = np.arange(10)

colors = ['blue', 'red', 'green']
color_gen = cycle(colors)

for i in range(1, 4):
    plt.scatter(x, i * x**2, label='Group %d' % i, color=next(color_gen))

plt.legend(loc='upper left')
plt.show()
```

```python
from itertools import cycle
x = np.arange(10)

colors = ['blue', 'red', 'green']
color_gen = cycle(colors)

for i in range(1, 4):
    plt.scatter(x, i * x**2, label='Group %d' % i, color=next(color_gen))

plt.legend(loc='upper left', scatterpoints=1)
plt.show()
```

```python
from itertools import cycle
x = np.arange(10)

colors = ['blue', 'red', 'green']
color_gen = cycle(colors)

for i in range(1, 4):
    plt.plot(x, i * x**2, label='Group %d' % i, marker='o')

plt.legend(loc='upper left')
plt.show()
```

```python
from itertools import cycle
x = np.arange(10)

colors = ['blue', 'red', 'green']
color_gen = cycle(colors)

for i in range(1, 4):
    plt.plot(x, i * x**2, label='Group %d' % i, marker='o')

plt.legend(loc='upper left', numpoints=2)
plt.show()
```
